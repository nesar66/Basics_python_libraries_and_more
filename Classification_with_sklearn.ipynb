```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_winequality=pd.read_csv("winequality.csv")
df_winequality.columns
```

```python
#checking for null values, total null value and remove any row that contains null value
df_winequality.isnull().values.any()
df_winequality.isnull().sum().sum()
df_winequality = df_winequality.dropna()
```

```python
df_winequality.head()
df_winequality.tail()
df_winequality.describe()
df_winequality=df_winequality.drop(['type'], axis=1)
```

```python
X=np.array(df_winequality.iloc[:,:-1])
y=np.array(df_winequality.quality).reshape(6463,)
```

```python
X[0:4,:]
```

```python
df_winequality.hist(bins=10, figsize=(25, 20))
# From the histogram we can say that the data is right skewed. Learn more about what right and left skew of the data means intuitively.
# For now just fit all the classifications method using model.fit and model.predict, then print accuracy score.
```

```python
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model_LogisticRegression=LogisticRegression()
model_SVM=SVC(probability=True)
model_Guasian_NB=GaussianNB()
model_Knn=KNeighborsClassifier()
model_Decision_trees=DecisionTreeClassifier()
model_Random_Forest=RandomForestClassifier()
```

```python
model_LogisticRegression.fit(X_train,y_train)
model_SVM.fit(X_train,y_train)
model_Guasian_NB.fit(X_train,y_train)
model_Knn.fit(X_train,y_train)
model_Decision_trees.fit(X_train,y_train)
model_Random_Forest.fit(X_train,y_train)
```

```python
model_LogisticRegression_result=model_LogisticRegression.predict(X_test)
model_SVM_result=model_SVM.predict(X_test)
model_Guasian_NB_result=model_Guasian_NB.predict(X_test)
model_Knn_result=model_Knn.predict(X_test)
model_Decision_trees_result=model_Decision_trees.predict(X_test)
model_Random_Forest_result=model_Random_Forest.predict(X_test)
```

```python
model_LogisticRegression_result_proba=model_LogisticRegression.predict_proba(X_test)
model_SVM_result_proba=model_SVM.predict_proba(X_test)
model_Guasian_NB_result_proba=model_Guasian_NB.predict_proba(X_test)
model_Knn_result_proba=model_Knn.predict_proba(X_test)
model_Decision_trees_result_proba=model_Decision_trees.predict_proba(X_test)
model_Random_Forest_result_proba=model_Random_Forest.predict_proba(X_test)
```

```python
print(model_LogisticRegression.score(X_test,y_test))
print(model_SVM.score(X_test,y_test))
print(model_Guasian_NB.score(X_test,y_test))
print(model_Knn.score(X_test,y_test))
print(model_Decision_trees.score(X_test,y_test))
print(model_Random_Forest.score(X_test,y_test))
```
