```python
#scatter plot
import numpy as np
import matplotlib.pyplot as plt
X_data=[23,24,45,67,78]
Y_data=[2345,5677,7766,5557,6790]
plt.scatter(X_data,Y_data,c="b", marker="*", s=150, alpha=.3)
plt.show()
```

```python
#lineplot
years= [2006+ x for x in range(16) ]
weights= [80,81,82,90,81,88,77,79,90,76,78,79,80,90,99,84]
plt.plot(years,weights, lw=5, linestyle="--") #lw means line width
plt.show()
```

```python
#bar chart for categorical variables
X=["c","c++", "Python", "Java"]
Y=[20,25,45,50]
plt.bar(X,Y,color="r", align="edge", width=.6, edgecolor="green", lw=6)
plt.show()
```

```python
#histogram
ages=np.random.normal(20,2.5,1000)
plt.hist(ages, bins=20)
plt.show()
```

```python
#pie chart
programming_language=["c","c++", "Python", "Java"]
votes=[20,25,45,50]
explodes=[.3,0,0,.3]
plt.pie(votes, labels=programming_language, explode=explodes, autopct="%.2f%%")
plt.show()
```

```python
#boxplot
heights=np.random.normal(172,8,300)
plt.boxplot(heights)
plt.show()
```

```python
years= [2014,2015,2016,2017,2018,2019,2020,2021]
weights= [55,56,62,61,71,72,74,75]
income_ticks= list(range(50,81,2))
plt.plot(years,weights, lw=5, linestyle="--") #lw means line width
plt.title("Income of John")
plt.xlabel("year")
plt.ylabel("Yearly Income of john")
plt.show()
```

```python
stock_a=[12,24,24,35,45,55]
stock_b=[22,44,4,35,45,75]
stock_c=[32,14,43,15,35,15]
plt.plot(stock_a, label="Company1")
plt.plot(stock_b,label="Company2")
plt.plot(stock_c,label="Company3")
plt.legend(loc="upper center")
plt.show()
```

```python
votes=[10,24,55,34,34]
people=["a","b","c","d","e"]
plt.pie(votes, labels=None) # plt.pie(votes, labels=people)
plt.legend(labels=people)
plt.show()
```

```python
x1,y1=np.random.random(100),np.random.random(100)
x2,y2=np.arange(100), np.random.random(100)
plt.figure(1)
plt.scatter(x1,y1)
plt.figure(2)
plt.plot(x2,y2)
plt.show()
```

```python
# sub figures
x=np.arange(100)
fig, axs= plt.subplots(2,2)
axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sin wave")
axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("Cos wave")
axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title("random wave")
axs[1,1].plot(x, np.log(x))
axs[1,1].set_title("log wave")
fig.suptitle("Four")
plt.tight_layout()
plt.show()
plt.savefig("nesar.png")
```

```python
#3d plots
ax=plt.axes(projection="3d")
x=np.random.random(100)
y=np.random.random(100)
z=np.random.random(100)
ax.scatter(x,y,z)
ax.set_title("3d scatter plot")
plt.show()
```
